```python
from bs4 import BeautifulSoup
import requests
```

```python
url = "https://globalrubbermarkets.com/?s=rubber"
```

```python
res = requests.get(url) # it opens up the connection and grabs the webpage
print(res)
```

```python
print(soup)
```

```python
container = soup.findAll("div",{"class":"td_module_16 td_module_wrap td-animation-stack"})
all_links = list()
for i in container:
    headlines = i.findAll("div",{"class":"item-details"})
    for j in headlines:
        date  = j.find("span",{"class":"td-post-date"})
        print(date.text)
        title = j.find("h3",{"class":"entry-title td-module-title"})
        print(title.text)
        summary = j.find("div",{"class":"td-excerpt"})
        print((summary.text).lstrip())
        link = j.find("a")
        print(link['href'])
        all_links.append(link['href'])
        print("\n")
```

```python
all_links
```

```python
for i in all_links:
    res_link = requests.get(i)
    link_soup = BeautifulSoup(res_link.text, 'lxml')
    content = link_soup.find('p')
    print(content.text)
    print("\n")
```
